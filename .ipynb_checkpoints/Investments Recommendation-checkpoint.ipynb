```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
import pickle
```

```python
df = pd.read_csv("process_2.csv", low_memory=False)
```

### we go to make clustering to this data to give investment recommendations for each segment of this data

```python
label_encoder = LabelEncoder()
# Function to label encode object columns
def label_encode_object_columns(df):
    object_columns = df.select_dtypes(include=['object']).columns
    for column in object_columns:
        df.loc[:, column] = label_encoder.fit_transform(df[column])  # Safe way to modify
    return df

# Apply label encoding
df_encoded = label_encode_object_columns(df)
# Scale numerical features
scaler = StandardScaler()
numeric_columns = df_encoded.select_dtypes(include=['number']).columns
df_scaled = scaler.fit_transform(df_encoded[numeric_columns])
# Convert back to DataFrame
df_scaled = pd.DataFrame(df_scaled, columns=numeric_columns)
```

```python
wcss = []
for i in range(1, 11):  # Trying cluster numbers from 1 to 10
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)  # Inertia = sum of squared distances to the nearest centroid

# Plot Elbow Method
#plt.figure(figsize=(8,5))
#plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
#plt.xlabel("Number of Clusters (k)")
#plt.ylabel("WCSS (Within-Cluster Sum of Squares)")
#plt.title("Elbow Method for Optimal k")
#plt.show()
```

```python
# Choose the optimal k (let’s say it's 4 from the elbow method)
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(df_scaled)
```

```python
# Add cluster labels to original dataframe
df['Cluster'] = kmeans.labels_

df.groupby(['Cluster']).mean()
```

```python
investment_recommendations = {
    0: [
        "Invest 40% in low-risk assets like government bonds or fixed deposits.",
        "Allocate 30% in moderate-risk assets such as index funds .",
        "Invest 20% in high-risk growth stocks .",
        "Consider investing 10% in alternative assets like real estate or REITs for diversification."
    ],
    1: [
        "Focus on reducing high-interest debt before investing. Prioritize debt repayment using the debt avalanche method.",
        "Once debt is reduced, allocate 50% of available funds to low-risk investments such as fixed deposits or savings accounts.",
        "Invest 30% in moderate-risk assets like regional ETFs or government bonds.",
        "Consider investing 20% in alternative assets like upskilling, education, or side business ventures to increase future earning potential."
    ],
    2: [
        "Allocate 30% to low-risk investments like government bonds or fixed deposits.",
        "Invest 40% in moderate-risk assets such as ETFs or blue-chip stocks.",
        "Invest 20% in higher-risk growth stocks or emerging market funds.",
        "Consider diversifying 10% in alternative investments such as real estate or cryptocurrency (e.g., Bitcoin or Ethereum)."
    ]
}
```

```python
# Now we assign investment recommendations to each cluter
# Assign investment recommendations
df["investment_recommendations"] = df["Cluster"].map(investment_recommendations)
```

```python
df.to_csv('Investment_recommendation.csv')
```

```python
df = pd.read_csv('Investment_recommendation.csv')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Convert all object type columns to numeric using LabelEncoder
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].apply(lambda x: ' '.join(x) if isinstance(x, list) else str(x))  # Convert lists to strings
    df[col] = label_encoder.fit_transform(df[col])  # Encode

# Step 2: Encode the target column 'Credit_Score'
df['investment_recommendations'] = label_encoder.fit_transform(df['investment_recommendations'])

# Step 3: Split the data into features (X) and target (y)
# Remove columns that should not be used as features
X = df.drop(['Unnamed: 0', 'ID', 'Customer_ID', 'Name', 'SSN', 'Not Specified',
       'Auto Loan', 'Mortgage Loan', 'Debt Consolidation Loan',
       'Personal Loan', 'Student Loan', 'Payday Loan', 'Home Equity Loan',
       'Credit-Builder Loan','investment_recommendations','Cluster'
            ], axis=1)
y = df['investment_recommendations']

# Step 4: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23
                                                   )
# Step 5: Initialize and train the Naïve Bayes model
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

# Step 6: Make predictions on the test set
y_pred = nb_model.predict(X_test)

# Step 7: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Display classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
